```python
# Importar librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar dataframes
df_consumption = pd.read_csv('data/Consumption.csv')
df_consumption['Fecha'] = pd.to_datetime(df_consumption['Fecha'])
df_consumption['Anio'] = df_consumption['Fecha'].dt.year
df_import = pd.read_csv('data/Import.csv')
df_import['Fecha'] = pd.to_datetime(df_import['Fecha'])
df_import['Anio'] = df_import['Fecha'].dt.year
```

Para este análisis exploratorio se optará por utilizar solamente las siguientes columnas:

- Gasolina regular
- Gasolina superior
- Diesel bajo azufre, Diesel ultra bajo azufre, Diesel alto azufre
- Gas licuado petroleo
- Fecha
- Total

de todos los conjutos.

```python
def descriptive_statics(df):
    columns_of_interest = ['Gasolina regular', 'Gasolina superior', 'Diesel bajo azufre',
                        'Diesel ultra bajo azufre', 'Diesel alto azufre', 'Gas licuado petroleo']
    subset_df = df[columns_of_interest]

    return subset_df.describe()

def plotting_distribution(subset_df, type_g : str):
    subset_df.hist(bins=20, figsize=(12, 8))
    plt.suptitle(f"Distribución de Variables de {type_g} de Combustible")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def plotting_behavior_gas_by_months(df, type_g : str):
    results_by_month = df.groupby(df['Fecha'].dt.month)['Total'].sum()
    plt.figure(figsize=(10, 6))
    results_by_month.plot(kind='bar', color='blue')
    plt.title(f'{type_g} de Gasolina por Mes')
    plt.xlabel('Mes')
    plt.ylabel(f'Total de {type_g}')
    plt.xticks(rotation=0)
    plt.show()
```

#### Análisis de Consumo

```python
# Explorar el comportamiento de las variables y su consumo (si esta normal o no)
d_statics_consum = descriptive_statics(df_consumption)
d_statics_consum
```

```python
plotting_distribution(d_statics_consum, "Consumo")
```

```python
# Meses en donde hay mas consumo de gas
plotting_behavior_gas_by_months(df_consumption, "Consumo")
```

```python
# Importaciones por año por tipo de combustibles
import matplotlib.pyplot as plt

fuel_types_to_plot = [
    "Gasolina superior",
    "Gasolina regular",
    "Diesel bajo azufre",
    "Diesel ultra bajo azufre",
    "Diesel alto azufre",
    "Gas licuado petroleo",
]

columns_exist = all(fuel_type in df_import.columns for fuel_type in fuel_types_to_plot)
columns_exist, df_import.columns if not columns_exist else None

yearly_import_data = df_import.groupby('Anio')[fuel_types_to_plot].sum().reset_index()

plt.figure(figsize=(14, 8))
for fuel_type in fuel_types_to_plot:
    plt.plot(yearly_import_data['Anio'], yearly_import_data[fuel_type], label=fuel_type)

plt.title('Importaciones por año por tipo de combustibles')
plt.xlabel('año')
plt.ylabel('Cantidad de importacion')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Comportamiento de los combustibles en los últimos años
fuel_types_to_plot = [
    "Gasolina superior",
    "Gasolina regular",
    "Diesel bajo azufre",
    "Diesel ultra bajo azufre",
    "Diesel alto azufre",
    "Gas licuado petroleo",
]

consumption_columns_exist = all(fuel_type in df_consumption.columns for fuel_type in fuel_types_to_plot)
consumption_columns_exist, df_consumption.columns if not consumption_columns_exist else None
yearly_consumption_data = df_consumption.groupby('Anio')[fuel_types_to_plot].sum().reset_index()

plt.figure(figsize=(14, 8))
for fuel_type in fuel_types_to_plot:
    plt.plot(yearly_consumption_data['Anio'], yearly_consumption_data[fuel_type], label=fuel_type)

plt.title('Consumo por año de los distintos tipos de combustibles ')
plt.xlabel('año')
plt.ylabel('Cantidad de consumo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Comportamiento de la importacion durante la pandemia
recent_yearly_import_data = df_import[df_import['Anio'] >= 2018].groupby('Anio')[fuel_types_to_plot].sum().reset_index()

plt.figure(figsize=(14, 8))
for fuel_type in fuel_types_to_plot:
    plt.plot(recent_yearly_import_data['Anio'], recent_yearly_import_data[fuel_type], label=fuel_type)

plt.title('Importaciones por año(2018 - presente) para los diferentes tipos de combustibles')
plt.xlabel('año')
plt.ylabel('Cantidad de importacion')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
prices_df = pd.read_csv('data/Precios_Promedio_Nacionales.csv')

prices_df.head()

prices_df['Fecha'] = pd.to_datetime(prices_df['Fecha'])


fuel_columns = ['Superior', 'Regular', 'Diesel', 'Bunker', 'Glp Cilindro 25Lbs.']
prices_df.plot(x='Fecha', y=fuel_columns, figsize=(15, 8), title='Comportamiento de los precios a lo largo del tiempo')
plt.ylabel('Precio')
plt.xlabel('Fecha')
plt.legend(title='Tipo')
plt.show()
```

```python
#Promedio de precio por mes año 2023
prices_2023_df = prices_df[prices_df['Anio'] == 2023]

mean_prices_2023_df = prices_2023_df.resample('M', on='Fecha').mean()[fuel_columns]

mean_prices_2023_df.plot(figsize=(15, 8), title='Precio promedio por mes 2023')
plt.ylabel('precio promedio')
plt.xlabel('Mes')
plt.legend(title='Tipo')
plt.show()
```
